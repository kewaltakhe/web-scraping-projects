```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from getpass import getpass
from time import sleep
from selenium.common.exceptions import NoSuchElementException
import json
```

```python
def get_brief_info(p):
    name = p.find_element('xpath','.//span[@dir="ltr"]/span').text
    profile_link = p.find_element('xpath','.//div[@class="display-flex"]//a').get_attribute('href')
    try:
        work = p.find_element('xpath','.//div[@class="mb1"]/div[2]/div[1]').text
    except NoSuchElementException:
        work = ''
    try:
        place = p.find_element('xpath','.//div[@class="mb1"]/div[2]/div[2]').text
    except NoSuchElementException:
        place = ''
    return (name,work,place,profile_link)
```

```python
##CONNECTING TO LinkedIn
driver = webdriver.Chrome()
driver.get('https://www.linkedin.com/login')
```

```python
##TARGET ACCOUNT -- go to connections
driver.get('https://www.linkedin.com/in/aman-s-5b5515108/')
driver.find_element('xpath','//ul[1]/li[1]/a[@class="ember-view"]').click()
```

```python
info_collected = []
while True:
    connection_profiles = driver.find_elements('xpath','.//div[@class="entity-result"]')
    for p in connection_profiles:
        result = get_brief_info(p)
        info_collected.append(result)
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight);')
    sleep(1)
    button = driver.find_element('xpath','//button[@aria-label="Next"]')
    
    if button.get_attribute('disabled') != None:
        break
    else:
        button.click()
        sleep(3)
```
